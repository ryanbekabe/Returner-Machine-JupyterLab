```python
from datetime import datetime
import itertools
#https://github.com/PZeeHiM/Plant-Seedlings-Classification/blob/master/utils_tools.py

def start_end_timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))

listDat = [0,1,2] #data perm harus urut
listDtt = (2,1,0) #data asli dibiarkan acak
perm = itertools.permutations(listDat)
c = 0
start_time = start_end_timer(None)
for i in list(perm):
    c = c + 1
    print(c,i)
    if listDtt == i:
        print('Sama indeks ke -',c)
        break;
print(c)
start_end_timer(start_time)
```
